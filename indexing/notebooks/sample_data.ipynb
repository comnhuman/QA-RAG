```python
from pathlib import Path

# 기준 디렉터리
base_dir = Path("../data")

# 허용되는 확장자 목록 (소문자로 비교)
allowed_exts = {
    ".pdf",
    ".docx",
    ".xlsx",
    ".pptx",
    ".md", ".markdown",
    ".adoc", ".asciidoc",
    ".html", ".htm", ".xhtml",
    ".csv",
    ".png",
    ".jpg", ".jpeg",
    ".tif", ".tiff",
    ".bmp",
    ".webp",
    ".vtt",   # WebVTT
    ".hwp",
}

# 이미 본 파일 이름들을 저장 (소문자로 통일)
seen_names = set()

for path in base_dir.rglob("*"):
    if not path.is_file():
        continue

    ext = path.suffix.lower()
    name = path.name.lower()  # 파일 이름(확장자 포함)

    # 1) 허용되지 않은 확장자이거나
    # 2) 이미 같은 이름의 파일을 본 경우
    #    → 현재(나중에 검색된) 파일 삭제
    if ext not in allowed_exts or name in seen_names:
        try:
            path.unlink()
            print(f"삭제됨: {path}")
        except Exception as e:
            print(f"삭제 실패: {path} ({e})")
    else:
        # 처음 보는 파일 이름이면 기록
        seen_names.add(name)
```

```python
from pathlib import Path

base_dir = Path("../data")  # 시작 디렉터리

target_name = "NIAT.pptx"

found = False
for path in base_dir.rglob(target_name):  # 하위 폴더까지 재귀적으로 검색
    print(f"찾음: {path}")
    found = True

if not found:
    print(f"파일을 찾을 수 없습니다: {target_name}")
```
